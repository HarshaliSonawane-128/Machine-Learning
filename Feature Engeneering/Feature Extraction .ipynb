### 1. Standardization

```python
import seaborn as sns
import pandas as pd 
```

```python
# load dataset

df = pd.read_csv('tips.csv')
```

```python
df.head()
```

```python
# histogram of total_bill

sns.histplot(df['total_bill'])
```

```python
# stadadization of total_bill and tip

from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()
```

```python
scalar .fit(df[['total_bill','tip']])
```

```python
a = scalar.transform(df[['total_bill','tip']])
a
```

```python
# or 
# scalar .fit_transform(df[['total_bill','tip']])
```

```python
## convert it into data frame 

a = pd.DataFrame(scalar.transform(df[['total_bill','tip']]),columns = ['total_bill' , 'tip'])
a.head()
```

```python
# concat two Dataframe 
# join two Dataframe

dff = pd.concat([a[['total_bill', 'tip']], df[['sex', 'smoker', 'day', 'time', 'size']]], axis=1)

dff.head()
```

```python
# histogramme of standardized total_bill

sns.histplot(a['total_bill'])
```

```python
## dirctly change the values of feature total_bill and tip in in to original dataset

df[['total_bill' ,'tip']] = scalar.transform(df[['total_bill','tip']])

df.head()
```
